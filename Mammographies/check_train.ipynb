```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pickle.load(open("C:/Users/Invite/Documents/INRIA/resnet_training/training_history.pkl", "rb")))
df.head()
```

```python
def ma(x, window):
    return [sum(x[i:i+window]) / window for i in range(len(x) - window + 1)]
```

```python
plt.plot(ma(df['train_acc'], window=5), label='Train Accuracy')
plt.plot(ma(df['val_acc'], window=5), label='Validation Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(ma(df['train_loss'], window=5), label='Train Loss')
plt.legend()
plt.show()
```

```python
plt.plot(ma(df['val_loss'], window=5), label='Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(ma(df['train_acc'] - df['val_acc'], window=5), label='Diff Accuracy')
plt.legend()
plt.show()
```
